```python
import numpy as np
import pandas as pd
```

```python
sim_tk = pd.read_excel('Simple Token.xlsx')
sim_lemma = pd.read_excel('Simple Lemma.xlsx')
spa_tk = pd.read_excel('Spacy Token.xlsx')
spa_lemma = pd.read_excel('Spacy Lemma.xlsx')
nltk_tk = pd.read_excel('Nltk Token.xlsx')
nltk_lemma = pd.read_excel('Nltk Lemma.xlsx')
```

```python
sim_tk['source'] = 'Simple Token'
sim_lemma['source'] = 'Simple Lemma'
spa_tk['source'] = 'Spacy Token'
spa_lemma['source'] = 'Spacy Lemma'
nltk_tk['source'] = 'Nltk Token'
nltk_lemma['source'] = 'Nltk Lemma'
```

```python
metric = sim_tk['metric']
```

```python
dfs = [sim_tk, sim_lemma, spa_tk, spa_lemma, nltk_tk, nltk_lemma]
combined = pd.concat(dfs, ignore_index=True)
combined
```

```python
combined = combined.dropna(subset=['overall_quality_score'], ignore_index=True)
combined
```

```python
com_df = combined.groupby(['metric'])[['overall_quality_score','feature_recommendation','source']].apply(lambda x: x)
com_df.to_excel('combined_features_summary.xlsx')
```

```python
idx = combined.groupby(['metric'])['overall_quality_score'].idxmax()
df = combined.loc[idx].reset_index(drop=True)
df = df.sort_values(by=['overall_quality_score'],ascending=False)
df.to_excel('best_features_summary.xlsx', index=False)
df
```

```python
import numpy as np

for i in range(len(metric)):
    current_metric_value = metric.iloc[i]

    # Use a dictionary to store scores with their corresponding dataframe names
    # Use .get() with a default of None to handle cases where a metric might not be in a dataframe
    scores_with_names = {
        'Simple Token': sim_tk[sim_tk['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not sim_tk[sim_tk['metric'] == current_metric_value].empty else np.nan,
        'Simple Lemma': sim_lemma[sim_lemma['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not sim_lemma[sim_lemma['metric'] == current_metric_value].empty else np.nan,
        'Spacy Token': spa_tk[spa_tk['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not spa_tk[spa_tk['metric'] == current_metric_value].empty else np.nan,
        'Spacy Lemma': spa_lemma[spa_lemma['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not spa_lemma[spa_lemma['metric'] == current_metric_value].empty else np.nan,
        'Nltk Token': nltk_tk[nltk_tk['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not nltk_tk[nltk_tk['metric'] == current_metric_value].empty else np.nan,
        'Nltk Lemma': nltk_lemma[nltk_lemma['metric'] == current_metric_value]['overall_quality_score'].iloc[0] if not nltk_lemma[nltk_lemma['metric'] == current_metric_value].empty else np.nan
    }

    # Filter out NaN values for comparison
    valid_scores = {name: score for name, score in scores_with_names.items() if not np.isnan(score)}

    if valid_scores:
        highest_score_name = max(valid_scores, key=valid_scores.get)
        highest_score = valid_scores[highest_score_name]
        print(f"Metric: '{current_metric_value}', Highest Overall Quality Score: {highest_score:.4f} (from {highest_score_name})")
    else:
        print(f"Metric: '{current_metric_value}', No valid scores found across dataframes.")
```

```python
import pandas as pd
import numpy as np

# Initialize an empty list to store the best rows
best_metric_rows = []

# Create a mapping from dataframe names (strings) to actual dataframe objects
df_mapping = {
    'Simple Token': sim_tk,
    'Simple Lemma': sim_lemma,
    'Spacy Token': spa_tk,
    'Spacy Lemma': spa_lemma,
    'Nltk Token': nltk_tk,
    'Nltk Lemma': nltk_lemma
}

for i in range(len(metric)):
    current_metric_value = metric.iloc[i]

    scores_with_names = {}
    for name, df_obj in df_mapping.items():
        # Check if the metric exists in the current dataframe before trying to access the score
        filtered_df = df_obj[df_obj['metric'] == current_metric_value]
        if not filtered_df.empty:
            scores_with_names[name] = filtered_df['overall_quality_score'].iloc[0]
        else:
            scores_with_names[name] = np.nan

    # Filter out NaN values for comparison
    valid_scores = {name: score for name, score in scores_with_names.items() if not np.isnan(score)}

    if valid_scores:
        highest_score_name = max(valid_scores, key=valid_scores.get)

        # Get the dataframe that corresponds to the highest score
        source_df = df_mapping[highest_score_name]

        # Extract the full row for the current metric from the source_df
        best_row = source_df[source_df['metric'] == current_metric_value].iloc[0].copy()

        # Add a new column to this row indicating the source dataframe
        best_row['source_dataframe'] = highest_score_name

        # Append the modified row to our list
        best_metric_rows.append(best_row)

# Create the new DataFrame from the list of best rows
overall_best_scores_df = pd.DataFrame(best_metric_rows)

# Display the new DataFrame
overall_best_scores_df
```
