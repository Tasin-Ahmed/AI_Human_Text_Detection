Extract text from DOCX file

```python
from docx import Document

file_path = 'AI Formatted.docx'

try:
    document = Document(file_path)
    full_text = []
    for para in document.paragraphs:
        # print(para.text)
        full_text.append(para.text)
    full_document_content = '\n'.join(full_text)
    print("Successfully read the DOCX file. The content is stored in 'full_document_content'.")
    print(f"First 500 characters of the document: {full_document_content[:500]}...")
except Exception as e:
    print(f"Error reading DOCX file: {e}")
```

Extracting Paragraphs

```python
import re

# Regex to find text enclosed in triple quotes (non-greedy match)
# This pattern looks for """ (three double quotes) followed by any characters (non-greedy) and then another """
# The 'DOTALL' flag allows '.' to match newlines as well.
pattern = r'"""(.*?)"""'
extracted_paragraphs = re.findall(pattern, full_document_content, re.DOTALL)

print(f"Found {len(extracted_paragraphs)} paragraphs enclosed in triple quotes.")
# Display the first few extracted paragraphs to verify
for i, para in enumerate(extracted_paragraphs[:3]):
    print(f"\n--- Paragraph {i+1} ---")
    print(para.strip()[:200]) # Print first 200 chars for brevity
```

Cleaning unexpected `newline` from text




```python
cleaned_paragraphs = []
for paragraph in extracted_paragraphs:
    cleaned_paragraph = paragraph.replace('\n', ' ').strip() # Replace newline with space and strip leading/trailing whitespace
    cleaned_paragraphs.append(cleaned_paragraph)

extracted_paragraphs = cleaned_paragraphs # Update the original list or use this new one for subsequent steps

print(f"Successfully cleaned {len(extracted_paragraphs)} paragraphs by removing newline characters.")
# Display the first few cleaned paragraphs to verify
for i, para in enumerate(extracted_paragraphs[:3]):
    print(f"\n--- Cleaned Paragraph {i+1} ---")
    print(para[:200]) # Print first 200 chars for brevity
```

Exporting XLSX file


```python
import pandas as pd

# Create a DataFrame from the cleaned extracted paragraphs
df_paragraphs = pd.DataFrame(extracted_paragraphs, columns=['Paragraph Content'])

# Save the DataFrame to an Excel file
output_excel_file = 'output_paragraphs_Bangla.xlsx'
df_paragraphs.to_excel(output_excel_file, index=True)

print(f"Successfully updated the DataFrame and saved {len(extracted_paragraphs)} cleaned paragraphs to '{output_excel_file}'.")
print("First 5 rows of the updated DataFrame:")
print(df_paragraphs.head())
```
